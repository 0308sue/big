```python
import json
import requests

api_key = #카카오 api

#위경도를 주소로
def lat_lon_to_addr(lat,lon):
    url = 'https://dapi.kakao.com/v2/local/geo/coord2regioncode.json?x={longitude}&y={latitude}'.format(longitude=lon,latitude=lat)
    headers = {"Authorization": "KakaoAK " + api_key}
    result = json.loads(str(requests.get(url, headers=headers).text))
    match_first = result['documents'][0]['address_name']
    return str(match_first)
```

```python
#주소를 위경도로
def addr_to_lat_lon(addr):
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query={address}'.format(address=addr)
    headers = {"Authorization": "KakaoAK " + api_key}
    result = json.loads(str(requests.get(url, headers=headers).text))
    match_first = result['documents'][0]['address']
    return float(match_first['y']),float(match_first['x']) 
```

```python
import pandas as pd
dfc = pd.read_excel('닭전염병_위경도수정.xlsx')
dfc = dfc.drop(['경도','위도'],axis=1)
dfc
```

```python
addr = dfc['주소'].to_list()
addr
```

```python
la = []
lo = []
for a in addr :
    try:
        lalo = addr_to_lat_lon(a)
        la.append(lalo[0])
        lo.append(lalo[1])
    except IndexError:
        la.append('-')
        lo.append('-')
#     print(a)
#     print(la)
dfc['위도'] = la
dfc['경도'] = lo
dfc
```
