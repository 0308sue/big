```python
from haversine import haversine
import warnings
warnings.filterwarnings(action='ignore')
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('가축전염병 통합데이터_220902_v1.xlsx',sheet_name='Sheet1')
df
```

```python
# country 컬럼을 선택합니다.
# 컬럼의 값과 조건을 비교합니다.
# 그 결과를 새로운 변수에 할당합니다.
chicken = df['품종'] == '닭'
# 사육개체수과 감염개체수
# 조건를 충족하는 데이터를 필터링하여 새로운 변수에 저장합니다.
dfc = df[chicken]

# 결과를 출력합니다.
dfc

# # dfcc = dfc.iloc[:,0:7]
# # dfccc = dfc.iloc[:,7:8]
# dfcc = dfc[['발생두수(마리)']]
# # dfccccc = dfc.iloc[:,14:17]
# # dfcccccc = dfc.iloc[:,19:23]
# # dfcc = pd.concat([dfccccc,dfcccccc],axis=1)
# dfcc.head()
```

```python
dead = pd.read_excel('가축전염병 통합데이터_220902_v1.xlsx',sheet_name='조류폐사체')
dead.head()
```

```python
result = pd.DataFrame(columns=['발생일자(가)','가축전염병명(가)','농장명(가)','농장소재지(가)','위도(가)','경도(가)','축종(가)','발생두수(가)','시료채취일자(가)','국가(가)','광역시도(가)','시군구(가)','읍면동(가)','읍면동 상세(가)','개체구분(가)','위도(가)','경도(가)','관측지코드(가)'])
listc = []
aa = 0
for c in range(len(dfc)):
    dfcr = dfc.loc[c]
    la = float(dfcr['위도'])
    lo = float(dfcr['경도'])
    chicken = (la,lo)  
    for a in range(len(dead)):
        deadr = dead.loc[a]
        laa = float(deadr['위도'])
        loa = float(deadr['경도'])
        ai = (laa,loa)
        # 거리 계산
        listc.append(haversine(chicken, ai, unit = 'km'))
        dfcli = dfcc.loc[c].to_list()
        dfali = dfaa.loc[a].to_list()
        dfdf = dfcli + dfali
        c = c+1
        bb = aa*cc
        i=bb-1
        result.loc[i]=dfdf
result
```
