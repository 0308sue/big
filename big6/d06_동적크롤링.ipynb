```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
path = 'D:\\chromdriver\\chromedriver_win32\\chromedriver.exe'
driver=webdriver.Chrome(path)
driver
```

```python
import urllib.request as req
from bs4 import BeautifulSoup as bs

driver.implicitly_wait(3)
driver.get('https://www.starbucks.co.kr/store/store_map.do')
target = driver.find_element(By.CLASS_NAME,'quickResultLstCon')
print(target.text)
# driver.quit()
```

```python
import time

time.sleep(2)
loca = driver.find_element(By.CLASS_NAME,'loca_search')
loca.click()
time.sleep(2)

# div.loca_step1 > div.loca_step1_cont > ul > li:nth-child(1)>a
f_link = driver.find_element(By.CSS_SELECTOR,'div.loca_step1 > div.loca_step1_cont > ul > li:nth-child(1)>a')
f_link.click()
time.sleep(2)

#mCSB_2_container > ul > li:nth-child(1)
f_link2= driver.find_element(By.CSS_SELECTOR,'#mCSB_2_container > ul > li:nth-child(1)>a')
f_link2.click()
time.sleep(2)

#mCSB_3_container > ul > li:nth-child(1)
f_link_all= driver.find_elements(By.CSS_SELECTOR,'#mCSB_3_container > ul > li')
time.sleep(2)
for s in f_link_all:
    print(s.text)
    
```

```python
shop_list=[]
for shop in f_link_all:
    lat = shop.get_attribute('data-lat')
    long = shop.get_attribute('data-long')
    name = shop.get_attribute('data-name')
    info = shop.find_element(By.TAG_NAME,'p')
    info_arr = info.text.split('\n')
    if(len(info_arr)==2):
        addr = info_arr[0]
        phone = info_arr[1]
        shop_list.append([name,lat,long,addr,phone])
    else : break
print(shop_list)
```
