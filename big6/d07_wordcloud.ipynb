```python
import pandas as pd
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from wordcloud import STOPWORDS,WordCloud
import nltk
```

```python
# nltk.download()
```

```python
all_data = pd.read_csv('riss_bigdata.csv',encoding='utf-8')
all_title = all_data['제목']
all_title
```

```python
stopWords = set(stopwords.words('english'))
lemma = WordNetLemmatizer()
words=[]

for title in all_title:
    EnWords = re.sub(r'[^a-zA-Z]',' ',str(title)) #영어가 아닌 것 ' '대체
    EnWordsToken = word_tokenize(EnWords.lower()) # 소문자로
    print(EnWordsToken)
    EnWordsTokenStop = [w for w in EnWordsToken if w not in stopWords]
    EnWordsTokenStopLemma = [lemma.lemmatize(w) for w in EnWordsTokenStop]
    words.append(EnWordsTokenStopLemma)
words
```

```python
from functools import reduce
#전처리가 끝난 words를 1 차원 list
words2 = list(reduce(lambda x,y:x+y,words))
words2
```

```python
from collections import Counter
count = Counter(words2)
count
```

```python
#상위 50개 단어 딕셔너리에 저장
word_count = dict()
for tag,counts in count.most_common(50):
    if(len(str(tag))>1):
        word_count[tag] = counts
word_count
```

```python
import matplotlib.pyplot as plt

sorted_Keys= sorted(word_count,key=word_count.get,reverse = True)
sorted_Values = sorted(word_count.values(),reverse=True)
plt.bar(range(len(word_count)),sorted_Values,align='center')
plt.xticks(range(len(word_count)),list(sorted_Keys),rotation='85')

plt.show()
```

```python
del word_count['big']
del word_count['data']
```

```python
stopwords = set(STOPWORDS)
wc = WordCloud(background_color='ivory',stopwords=stopwords,width=800,height=600)
cloud = wc.generate_from_frequencies(word_count)
plt.figure(figsize=(8,8))
plt.imshow(cloud)
plt.axis('off')
plt.show()
cloud.to_file('k_wordCloud.jpg')
```
