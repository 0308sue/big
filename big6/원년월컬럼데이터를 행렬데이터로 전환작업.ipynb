```python
## 평균 컬럼(8,9,10번째) 은 삭제함
df2 = df.drop(df.columns[[0,8,9,10]],axis=1)
df2.head()
```

```python
##* Table Join
##   - right  오른쪽 테이블의 key값 기준으로 데이터를 생성하고, 값이 없는 대로 생성

df7=pd.merge(df5,df3, how='right', on='구분')
df7.to_excel('경기도_겨울철새202110202203.xlsx')
```
