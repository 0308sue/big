```python
# country 컬럼을 선택합니다.
# 컬럼의 값과 조건을 비교합니다.
# 그 결과를 새로운 변수에 할당합니다.
chicken = df['축종'] == '닭'
# 사육개체수과 감염개체수
# 조건를 충족하는 데이터를 필터링하여 새로운 변수에 저장합니다.
dfc = df[chicken]
dfc = dfc.drop(['종식일'],axis=1)
# 결과를 출력합니다.


# dfcc = dfc.iloc[:,0:5]
# dfccc = dfc.iloc[:,5:14]
# dfcc = dfc[['발생두수(마리)']]
# # dfccccc = dfc.iloc[:,14:17]
# # dfcccccc = dfc.iloc[:,19:23]
# dfc = pd.concat([dfcc,dfccc],axis=1,)
dfc
# dfcc.head()
```

```python
result = pd.DataFrame(columns=['발생일자(가)','가축전염병명(가)','국가(가)','광역시도(가)','시군구(가)','읍면동(가)','리(가)','읍면동 상세(가)','축종(가)','발생두수(마리)(가)','위도(가)','경도(가)','관측지코드(가)','년월(폐)','국가(폐)','광역시도(폐)','시군구(폐)','서식지명(폐)','개체수(폐)','위도(폐)','경도(폐)','관측지코드(폐)'])
listc = []
aa = 0
for c in range(len(dfc)):
    dfcr = dfc.loc[c]
    la = float(dfcr['위도'])
    lo = float(dfcr['경도'])
    chicken = (la,lo)  
    cc = c+1
    for a in range(len(dead)):
        deadr = dead.loc[a]
        laa = float(deadr['위도'])
        loa = float(deadr['경도'])
        ai = (laa,loa)
        # 거리 계산
        listc.append(haversine(chicken, ai, unit = 'km'))
        dfcli = dfcr.to_list()
        dfali = deadr.to_list()
        dfdf = dfcli + dfali  
        aa=aa+1
        bb = aa*cc
        i=bb-1
        result.loc[i]=dfdf
result
```

```python
result['거리(KM)'] = listc
result
```

```python
result = pd.read_excel('aa.xlsx')
result
result = result.drop(['Unnamed: 0'],axis=1)
result
```

```python
from datetime import datetime
date_diff_list = []
happen = result['발생일자(가)']
collect = result['년월(폐)']

for h in range(len(happen)):
    happenr = happen.loc[h]
    collectr = collect.loc[h]
    
    happend = datetime.strptime(happenr, "%Y-%m-%d")
    collectd = datetime.strptime(collectr, "%Y-%m")
    
    date_diff = collectd - happend
    date_diff = date_diff*-1
    date_diff_list.append(date_diff)

result['시료채취일_발생일_비교'] = date_diff_list
result
```
