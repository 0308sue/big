```python
import pandas as pd
from datetime import datetime
from pandas import json_normalize
```

```python
address_df = pd.read_csv('주소.csv', encoding='utf8')
address_df['리'] = address_df['리'].fillna('')
address_df
```

```python
user_key = '00170112-DCDA-3D9C-BCEA-FD32C805A371'
domain = 'localhost'

list = []

for address in address_df.loc[:,'리']:

    # 읍면동 주소 요청
    # api_url = 'https://api.vworld.kr/req/data?service=data&request=GetFeature&data=LT_C_ADEMD_INFO&key='+user_key+'&domain='+domain+'&attrFilter=full_nm:=:'+address
    
    # 리주소 요청
    api_url = 'http://api.vworld.kr/req/data?service=data&request=GetFeature&data=LT_C_ADRI_INFO&key='+user_key+'&domain='+domain+'&attrFilter=li_kor_nm:=:'+address

    # url 불러오기
    response = requests.get(api_url)

    #데이터 값 출력해보기
    contents = response.text

    #문자열을 json으로 변경
    json_ob = json.loads(contents)
    
    if json_ob['response']['status'] =='OK':
        latitude = json_ob['response']['result']['featureCollection']['bbox'][1]
        longitude = json_ob['response']['result']['featureCollection']['bbox'][0]
        list.append([address, latitude, longitude])
    else:
        list.append([address, ' ', ' '])
    
result = pd.DataFrame(list)
result
```

```python
result.to_csv('주소 위경도.csv', encoding='CP949')
```
