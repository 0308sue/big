```python
#p278 비지도 학습
from sklearn import cluster,datasets
import matplotlib.pyplot as plt
iris = datasets.load_iris()
x = iris.data[:,2:]
y = iris.target
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.axis([0,x[:,0].max()+0.2,0,x[:,1].max()+0.2])
plt.plot(x[:,0],x[:,1],'bo')
```

```python
plt.scatter(x[:,0],x[:,1],c=y,cmap='viridis',edgecolors='k')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.axis([0,x[:,0].max()+0.2,0,x[:,1].max()+0.2])
plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np
X = np.array([[1,5],[3,4],[2,3],[5,3],[6,2],[5,1]])
kmeans = KMeans(n_clusters=2,random_state=0).fit(X)
# kmeans = kmeans.fit(x)
print(kmeans.labels_)
print(kmeans.cluster_centers_)
```

```python
y = kmeans.labels_
ycenter = kmeans.cluster_centers_
plt.scatter(X[:,0],X[:,1],c=y,cmap='viridis',edgecolors='k')
plt.scatter(ycenter[:,0],ycenter[:,1],c=[0,1],s=200,cmap='viridis',edgecolors='k')
```

```python
X = np.array([[1,5],[3,4],[2,3],[5,3],[6,2],[5,1]])
kmeans = KMeans(n_clusters=2,random_state=0).fit(X)
x = np.array([[1,4],[4,2]])
y_pred = kmeans.predict(x)
print(y_pred)
plt.scatter(X[:,0],X[:,1],c=y,cmap='viridis',edgecolors='k')
plt.scatter(x[:,0],x[:,1],c=y_pred,s=200,cmap='magma',marker='^',edgecolor='k')
```

```python
X = np.array([[1,5],[3,8],[3,5],[10,13],[11,14],[5,3],[7,2],[5,1]])
kmeans = KMeans(n_clusters=3,random_state=0).fit(X)
y=kmeans.labels_
print(y)
print(kmeans.cluster_centers_)
x = np.array([[1,4],[6,5],[8,10]])
y_pred = kmeans.predict(x)
plt.scatter(X[:,0],X[:,1],c=y,cmap='viridis',edgecolor='k')
plt.scatter(x[:,0],x[:,1],c=y_pred,s=200,cmap='magma',marker='^',edgecolor='k')
plt.axis([0,12,0,15])
plt.show()
```

```python
X_iris = iris.data
y_iris = iris.target

k_means = cluster.KMeans(n_clusters=3)
k_means.fit(X_iris)
print(k_means.labels_[::10])
print(y_iris[::10])
```
