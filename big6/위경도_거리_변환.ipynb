```python
# country 컬럼을 선택합니다.
# 컬럼의 값과 조건을 비교합니다.
# 그 결과를 새로운 변수에 할당합니다.
chicken = df['축종(품종)'] == '닭'

# 조건를 충족하는 데이터를 필터링하여 새로운 변수에 저장합니다.
dfc = df[chicken]

# 결과를 출력합니다.
dfc

dfcc = dfc.iloc[:,0:7]
dfccc = dfc.iloc[:,7:8]
dfcc = pd.concat([dfccc,dfcc],axis=1)
dfcc
```

```python
dfa = pd.read_excel('야생조류 고병원성AI 검출현황 21년01월 22년03월(월별, 위경도, 건수)_위경도수정.xlsx')
dfa
dfaa = dfa.iloc[:,1:5]
dfaa2 = dfa.iloc[:,8:11]
dfaa = pd.concat([dfaa,dfaa2],axis=1)
dfaa
```

```python
result = pd.DataFrame(columns=['발생일자(가)','가축전염병명(가)','광역시도(가)','시군구(가)','읍면동(가)','리(가)','위도(가)','경도(가)','시료채취일자(야)','시도(야)','시군(야)','주소(야)','시료(야)','위도(야)','경도(야)'])
listc = []
for c in range(len(dfcc)):
    dfccr = dfcc.loc[c]
    la = float(dfccr['위도'])
    lo = float(dfccr['경도'])
    chicken = (la,lo)  
    for a in range(len(dfaa)):
        dfaar = dfaa.loc[a]
        laa = float(dfaar['위도'])
        loa = float(dfaar['경도'])
        ai = (laa,loa)
        # 거리 계산
        listc.append(haversine(chicken, ai, unit = 'km'))
        dfcli = dfcc.loc[c].to_list()
        dfali = dfaa.loc[a].to_list()
        dfdf = dfcli + dfali
        c = c+1
        a = a+1
        i = c*a-1
        result.loc[i]=dfdf
result
```
