```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()

input_data = [[270,0],[260,0],[250,0],[220,1],[230,1],[240,1]]
fruits=[0,0,0,1,1,1]
clf =  clf.fit(input_data,fruits)
clf
```

```python
print(clf.predict([[250,1]]))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,17))
tree.plot_tree(clf,impurity=True,filled=True,rounded=True)
```

```python
import graphviz
dot_data = tree.export_graphviz(clf,out_file=None)
graph = graphviz.Source(dot_data)
graph.render()

dot_data = tree.export_graphviz(clf,impurity=True,filled=True,rounded=True,special_characters=True)
graph = graphviz.Source(dot_data)
graph
```
