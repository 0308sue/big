```python
X_train = X_train.iloc[:,1:]
X_train
```

```python
X_test_id = X_test['cust_id']
X_test_id
```

```python
X_test = X_test.iloc[:,1:] #cust_id 삭제
X_test
```

```python
y_train = y_train.iloc[:,-1:]
y_train
```

```python
from sklearn.preprocessing import LabelEncoder

X_train.loc[:,['주구매상품','주구매지점']] = \
X_train.loc[:,['주구매상품','주구매지점']].apply(LabelEncoder().fit_transform)
```

```python
# X_test 적용
X_test.loc[:,['주구매상품','주구매지점']] = \
X_test.loc[:,['주구매상품','주구매지점']].apply(LabelEncoder().fit_transform)
```

```python
X_train.isnull().sum()
```

```python
#환불금액 NaN을 0으로
# X_train['환불금액'] = X_train['환불금액'].fillna(0)
# X_train.isnull().sum()

#loc() 사용
X_train.loc[:,['환불금액']] = X_train.loc[:,['환불금액']].fillna(0)
X_train.isnull().sum()
```

```python
# test set 적용
X_test['환불금액'] = X_test['환불금액'].fillna(0)
X_train.isnull().sum()
```

```python
#로지스틱 회귀분석 ==> 분류
from sklearn.linear_model import LogisticRegression

md = LogisticRegression(C=1,random_state=0)
md.fit(X_train,y_train)
print('xscore: ',md.score(X_train,y_train))
```

```python
#predict_proba : 해당 레이블로 분류될 확률 값
pred = md.predict_proba(X_test)
pred = pd.DataFrame(pred)
pred = pred.iloc[:,1]
pred
```

```python
ans = pd.concat([X_test_id,pred],axis = 1)
ans
```

```python
#K-최근접 이웃 알고리즘
from sklearn.neighbors import KNeighborsClassifier
md = KNeighborsClassifier(n_neighbors=10,metric='euclidean')
md.fit(X_train,y_train)
print('score:',md.score(X_train,y_train))
```

```python
pred = md.predict_proba(X_test)
pred = pd.DataFrame(pred)
pred = pred.iloc[:,1]
pred
```
