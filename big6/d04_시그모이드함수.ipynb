```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(wx):
    return 1/(1+np.exp(-wx))

wx = np.arange(-8.0,9.0,0.1)
y = sigmoid(wx)

plt.plot(wx,y,'b')
plt.grid()
plt.xlabel('z',size=10)
plt.ylabel('(z)',size=10)
plt.title('Logistic|sigmoid function')
plt.show()
```

```python
#승산(odds) : 확률에서 시작되어 y=1 일 성공이라 가정할 때 실패(1-p)에 대한 성공(P)알 확률 비를 의미
#odds의 문제점 : p가 1에 가까울 수록 승산이 치솟음
#0 <= odds < 무한대
#로짓 변환 그래프
#입력 0~1 사이에 대해서 출력값은 -무한대 ~무한대를 가짐

def odds(p):
    return p/(1-p)

#확률 p : 0 ~ 1
p = np.arange(0.001,0.999,0.001)
z = [odds(n) for n in p]

plt.plot(p,z)
plt.ylim([0,200.0])
plt.xlabel("p")
plt.ylabel("odds")
plt.show()
```

```python
def logit1(p):
    return np.log(p/(1-p))

p = np.arange(0.001,0.999,0.001)
z = [logit1(n) for n in p]

plt.plot(p,z)
plt.xlabel("p")
plt.ylabel("logit")
plt.show()
```
